## **Pertanyaan 2** : Bandingkan tren dari 5 negara teratas yang menyumbang imigran ke Kanada

**Membuat line plot untuk membandingkan tren dari 5 negara teratas yang menyumbang imigran ke Kanada**

```python
top5_index = df_can['Total'].sort_values(ascending=False).head().index
top5_df = df_can.loc[top5_index, years]
top5_df = top5_df.T
```

**Membuat line plot**

```python
top5_df.plot(figsize=(15, 8));
plt.title("Imgiran dari 5 negara teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

### Area Plot

**Membuat area plot untuk membandingkan tren dari 5 negara teratas yang menyumbang imigran ke Kanada**

```python
top5_df.plot(kind='area',
             stacked=False,
             figsize=(15, 8));
plt.title("Imgiran dari 5 negara teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

### Histogram

**Menyimpan data 2013**

```python
df_2013 = df_can[2013]
```

**Numpy**

```python
count, bin_edges = np.histogram(df_2013, bins=10)

print(count)
print(bin_edges)
```

**Membuat histogram dari data Imigran 2013**

```python
df_2013.plot(kind='hist',
             bins=10,
             xticks=bin_edges,
             figsize=(8,5));
plt.title("Histogram dari data Imigran 2013");
plt.xlabel("Jumlah Imigran");
plt.ylabel("Jumlah Negara");
```

## **Pertanyaan 3** : Bagaimana distribusi dari Imigran Denmark, Norwegia dan Swedia ?

### Pie Chart

**Meringkas data imigrasi berdasarkan Continent**

```python
df_continents = df_can.groupby('Continent').sum()
df_continents
```

**Membuat pie chart data imigrasi ke Kanada berdasarkan Continent**

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct="%.2f%%",
                            startangle=90,
                            shadow=True);
```

**Memodifikasi untuk meningkatkan visual**

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continents['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct="%.1f%%",
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.15,
                            explode=explode_list);

plt.legend(labels=df_continents.index);
plt.title("Data Imigrasi sesuai Continent", y=1.12);
plt.axis('equal');
```
